```python
import xml.etree.cElementTree as ET
tree = ET.parse('nchltAux1_xho.xml')
root = tree.getroot()
st = []
for recording in root.iter('recording'):
    WAV_FILE = recording.get('audio')
    ID = WAV_FILE[38:41]
    UTT_ID = WAV_FILE[43:47]
    ID_UTT = ID + '_' + UTT_ID
    ID_UT  = ID_UTT + ' ' + ID
    st.append(ID_UT)
st.sort()
x = 0        
with open('utt2spk.txt', 'w') as writer3:
    for a in st:
        writer3.write(st[x])
        writer3.write('\n')
        x = x +1
```
