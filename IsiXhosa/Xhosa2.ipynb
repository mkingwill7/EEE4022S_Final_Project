```python
import xml.etree.cElementTree as ET
tree = ET.parse('nchltAux1_xho.xml')
root = tree.getroot()
st = []
for recording in root.iter('recording'):
    WAV_FILE = recording.get('audio')
    ID = WAV_FILE[38:41]
    UTT_ID = WAV_FILE[43:47]
    ID_UTT = ID + '_' + UTT_ID
    ID_UT  = ID_UTT + ' ' + ID
    st.append(ID_UT)
st.sort()

with open('utt2spk_test_reduc.txt','w') as writer:
    for x in range(1077):
        writer.write(st[x])
        writer.write('\n')
        


with open('utt2spk_train_reduc.txt','w') as writer:
    for y in range(5381):
        writer.write(st[y+1077])
        writer.write('\n')
        
```

```python
WAV_ARR = []
for recording in root.iter('recording'):
        WAV_FILE = recording.get('audio')
        ID = WAV_FILE[38:41]
        UTT_ID = WAV_FILE[43:47]
        IDD = ID +'_'+UTT_ID
        WAV_ARR.append(IDD)
tot = []        
i = 0;
for trans in root.iter('orth'):
    totID = WAV_ARR[i] + ' ' + trans.text
    tot.append(totID)
    i = i +1
tot.sort()
#print(tot[0])
with open('text_test_reduc.txt','w') as writer:
    for x in range(1077):
        writer.write(tot[x])
        writer.write('\n')
        


with open('train_train_reduc.txt','w') as writer:
    for y in range(5381):
        writer.write(tot[y+4801])
        writer.write('\n')
        
with open('corpus_reduc.txt','w') as writer:
    for z in range(6458):
        writer.write(tot[z][9:])
        writer.write('\n')
```
