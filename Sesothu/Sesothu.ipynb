```python
import xml.etree.cElementTree as ET
tree = ET.parse('nchltAux1_sot.xml')
root = tree.getroot()
st = []
for recording in root.iter('recording'):
    WAV_FILE = recording.get('audio')
    ID = WAV_FILE[38:41]
    UTT_ID = WAV_FILE[43:47]
    ID_UTT = ID + '_' + UTT_ID
    ID_UT  = ID_UTT + ' ' + ID
    st.append(ID_UT)
st.sort()

with open('utt2spk_test.txt','w') as writer:
    for x in range(4325):
        writer.write(st[x])
        writer.write('\n')
        


with open('utt2spk_train.txt','w') as writer:
    for y in range(64274):
        writer.write(st[y+4325])
        writer.write('\n')
```

```python
lexi =  open('lexicon.txt','r') 
lex = []
for x in lexi:
    lex.append(x.split())
lexi.close()
newlex = []
for i in range(len(lex)):
    newlex.append(lex[i][1:])
phones = []
for x in range(len(newlex)):
    for y in range(len(newlex[x])):
        if newlex[x][y] in phones:
            continue
        else:
            phones.append(newlex[x][y])
with open('nonsilence_phones.txt','w') as writer:
    for a in phones:
        writer.write(a)
        writer.write('\n')
```
